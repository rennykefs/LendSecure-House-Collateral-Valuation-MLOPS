```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df=pd.read_csv("C:\\Users\\Jones Mbela\\Desktop\\RENNY\\AI AND ML\\Lend Secure\\Data\\cleaned_real_estate_data.csv")
current_year= datetime.now().year
```

```python
#Building Age Feature
df["building_age"]= current_year - df["building_construction_year"]
df=df.drop(columns=["building_construction_year"])
```

```python
#Area per Room Feature(log)
df["log_apartment_total_area"]=np.log1p(df["apartment_total_area"])
df=df.drop(columns=["apartment_total_area"])
```

```python
#Country one hot encoding Feature
df=pd.get_dummies(df, columns=["country"], drop_first=True)
```

```python
#Location frequency encoding Feature
location_freq= df["location"].value_counts(normalize=True)
df["location_freq"]= df["location"].map(location_freq)
df=df.drop(columns=["location"])
```

```python
#Median imputation for any remaining missing values
num_features=[
    "building_total_floors",
    "apartment_rooms",
    "apartment_bathrooms",
    "building_age",
    "log_apartment_total_area",

]
for col in num_features:
    df[col]= df[col].fillna(df[col].median())
```
